```python
import pandas as pd
```

```python
df = pd.read_csv("questions_orig.csv")
df['llm_labels'] = df['llm_labels'].str.replace("General Wildfire behavior: causes, spread, fire science, smoke behavior, seasons, risk factors.", "General Wildfire behavior")
df['Question_Type'] = df['llm_labels']
df = df.drop(columns=["llm_labels",'mismatch_'])

# now update with questions.csv where rows overlap
df_new = pd.read_csv("questions.csv")

merge_keys = ['Source', 'Question', 'Answer', 'City', 'State', 'Question_Type']
update_cols = ['Location Dependent', 'llm_yesno', ]

df = df.merge(
    df_new[merge_keys + update_cols],
    on=merge_keys,
    how='left',
    suffixes=('_orig', '_new')
)

for col in update_cols:
    df[col] = df[f'{col}_new'].combine_first(df[f'{col}_orig'])
    df.drop(columns=[f'{col}_orig', f'{col}_new'], inplace=True)
df['Location Dependent'] = df['llm_yesno']
df = df.drop(columns=["llm_yesno"])
```
