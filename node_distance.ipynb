```python
def hierarchy_distance(h1, h2):
    set1 = set(h1)
    set2 = set(h2)
    common = set1 & set2
    if not common:
        return float("inf")
    lca = None
    for h in h1:
        if h in common:
            lca = h
            break
    d1 = h1.index(lca)
    d2 = h2.index(lca)
    return d1 + d2
```

```python
denver = ["Denver", "Denver County", "Colorado", "West", "United States"]
aspen  = ["Aspen", "Pitkin County", "Colorado", "West", "United States"]

print(hierarchy_distance(denver, aspen))  # SHould be 4
```
